```python
import matplotlib.pyplot as mp
import numpy as np
```

```python
# a) part
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
```

```python
mp.scatter(x,y)
mp.show()
```

```python
# b) part
x, y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))

mu = np.array([0, 1])
sigma = np.array([[1, 0.25], [0.25, 1]])
```

```python
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x
pos[:, :, 1] = y
inv_sigma = np.linalg.inv(sigma)
det_sigma = np.linalg.det(sigma)
Z = np.exp(-0.5 * np.sum((pos - mu) @ inv_sigma * (pos - mu), axis=2)) / (2 * np.pi * np.sqrt(det_sigma))
```

```python
fig, ax = mp.subplots()
ax.contour(x, y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('2D Gaussian Distribution')
mp.show()

```
